```python
import matplotlib.pyplot as plt

c_output = [[],[],[],[]]
with open('/home/domenico/Desktop/Unipv/Magistrale/Thesis/HSS-on-RiscV/src/python/output_risc.txt') as f: #change with your path containing results from python and from the quantized net
    lines = f.readlines()
    for line in lines:
        tmp = line.split()
        c_output[0].append(float((tmp[0])))
        c_output[1].append(float((tmp[1])))
        c_output[2].append(float((tmp[2])))
        c_output[3].append(float((tmp[3])))

python_output = [[],[],[],[]]
with open('/home/domenico/Desktop/Unipv/Magistrale/Thesis/HSS-on-RiscV/src/python/python_result.txt') as f: #change with your path containing results from python and from the quantized net
    lines = f.readlines()
    for line in lines:
        tmp = line.split()
        python_output[0].append(float((tmp[0])))
        python_output[1].append(float((tmp[1])))
        python_output[2].append(float((tmp[2])))
        python_output[3].append(float((tmp[3])))
```

```python
fig, axs = plt.subplots(4, 1, figsize=(10, 10))
for i in range(4):
    axs[i].plot(c_output[i], c='blue', alpha=0.4)
    axs[i].plot(python_output[i], c='red', alpha=0.4)
    axs[i].set_title(f'Channel {i}')
    axs[i].set_xlabel('Samples')
    axs[i].set_ylabel('Amplitude')
    axs[i].legend(['FP model', 'Original model'])
plt.tight_layout()
plt.show()
```
