```python
import numpy as np
from tensorflow.keras import layers, Input, Model
import tensorflow as tf
from models import *
import os 

# Go to paremeters directory 
os.chdir('../../data/weights/') 

# Load the kernel weights 
init_enc_0_conv_relu_0 = np.load('enc_0_conv_relu_0.npy')
init_enc_0_conv_relu_1 = np.load('enc_0_conv_relu_1.npy')
init_enc_1_conv_relu_0 = np.load('enc_1_conv_relu_0.npy')
init_enc_1_conv_relu_1 = np.load('enc_1_conv_relu_1.npy')
init_enc_2_conv_relu_0 = np.load('enc_2_conv_relu_0.npy')
init_enc_2_conv_relu_1 = np.load('enc_2_conv_relu_1.npy')
init_enc_3_conv_relu_0 = np.load('enc_3_conv_relu_0.npy')
init_enc_3_conv_relu_1 = np.load('enc_3_conv_relu_1.npy')
init_cen_0 = np.load('central_conv_relu_0.npy')
init_cen_1 = np.load('central_conv_relu_1.npy')
init_dec_0_up = np.load('dec_0_up_conv_relu.npy')
init_dec_0_up_conv_relu_0 = np.load('dec_0_conv_relu_0.npy')
init_dec_0_up_conv_relu_1 = np.load('dec_0_conv_relu_1.npy')
init_dec_1_up = np.load('dec_1_up_conv_relu.npy')
init_dec_1_up_conv_relu_0 = np.load('dec_1_conv_relu_0.npy')
init_dec_1_up_conv_relu_1 = np.load('dec_1_conv_relu_1.npy')
init_dec_2_up = np.load('dec_2_up_conv_relu.npy')
init_dec_2_up_conv_relu_0 = np.load('dec_2_conv_relu_0.npy')
init_dec_2_up_conv_relu_1 = np.load('dec_2_conv_relu_1.npy')
init_dec_3_up = np.load('dec_3_up_conv_relu.npy')
init_dec_3_up_conv_relu_0 = np.load('dec_3_conv_relu_0.npy')
init_dec_3_up_conv_relu_1 = np.load('dec_3_conv_relu_1.npy')
init_final_conv = np.load('final_conv.npy')

kernel_initializer_enc_0_0 = tf.keras.initializers.constant(init_enc_0_conv_relu_0)
kernel_initializer_enc_0_1 = tf.keras.initializers.constant(init_enc_0_conv_relu_1)
kernel_initializer_enc_1_0 = tf.keras.initializers.constant(init_enc_1_conv_relu_0)
kernel_initializer_enc_1_1 = tf.keras.initializers.constant(init_enc_1_conv_relu_1)
kernel_initializer_enc_2_0 = tf.keras.initializers.constant(init_enc_2_conv_relu_0)
kernel_initializer_enc_2_1 = tf.keras.initializers.constant(init_enc_2_conv_relu_1)
kernel_initializer_enc_3_0 = tf.keras.initializers.constant(init_enc_3_conv_relu_0)
kernel_initializer_enc_3_1 = tf.keras.initializers.constant(init_enc_3_conv_relu_1)
kernel_initializer_cen_0 = tf.keras.initializers.constant(init_cen_0)
kernel_initializer_cen_1 = tf.keras.initializers.constant(init_cen_1)
kernel_initializer_dec_0_up = tf.keras.initializers.constant(init_dec_0_up)
kernel_initializer_dec_0_0 = tf.keras.initializers.constant(init_dec_0_up_conv_relu_0)
kernel_initializer_dec_0_1 = tf.keras.initializers.constant(init_dec_0_up_conv_relu_1)
kernel_initializer_dec_1_up = tf.keras.initializers.constant(init_dec_1_up)
kernel_initializer_dec_1_0 = tf.keras.initializers.constant(init_dec_1_up_conv_relu_0)
kernel_initializer_dec_1_1 = tf.keras.initializers.constant(init_dec_1_up_conv_relu_1)
kernel_initializer_dec_2_up = tf.keras.initializers.constant(init_dec_2_up)
kernel_initializer_dec_2_0 = tf.keras.initializers.constant(init_dec_2_up_conv_relu_0)
kernel_initializer_dec_2_1 = tf.keras.initializers.constant(init_dec_2_up_conv_relu_1)
kernel_initializer_dec_3_up = tf.keras.initializers.constant(init_dec_3_up)
kernel_initializer_dec_3_0 = tf.keras.initializers.constant(init_dec_3_up_conv_relu_0)
kernel_initializer_dec_3_1 = tf.keras.initializers.constant(init_dec_3_up_conv_relu_1)
kernel_initializer_final = tf.keras.initializers.constant(init_final_conv)
```

```python
# Second convolution output 
second_conv_output = layers.Conv1D(filters=8,
                                kernel_size=3,
                                strides=1,
                                activation="relu",
                                # input_shape = X_0_0.shape,
                                padding="same",
                                use_bias=False,
                                kernel_initializer=kernel_initializer_enc_0_1)(first_conv_output)
```

```python
# From tensor to numpy to plot
second_conv_output_npy = second_conv_output.numpy()

# Same plot as before but with the output of the second convolutional layer
fig, axs = plt.subplots(8, 1, figsize=(10, 10))
for i in range(8):
        axs[i].plot(second_conv_output[0, :, i])
        axs[i].set_title(f'Channel {i}')
        axs[i].set_xlabel('Samples')
        axs[i].set_ylabel('Amplitude')
```

```python
# Store output in a txt file
second_conv_output_npy = second_conv_output.numpy()
np.savetxt('second_conv_output.txt', second_conv_output_npy[0, :, :])
```

```python
# MaxPooling output 
max_pool_out = layers.MaxPooling1D(pool_size=2, strides=2, padding="same")(second_conv_output) 
```

```python
enc1_conv_0_output = layers.Conv1D(filters=16,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_enc_1_0)(max_pool_out)
enc1_conv_1_output = layers.Conv1D(filters=16,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_enc_1_1)(enc1_conv_0_output)
max_pool_out_11 = layers.MaxPooling1D(pool_size=2, strides=2, padding="same")(enc1_conv_1_output)
enc2_conv_0_output = layers.Conv1D(filters=32,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_enc_2_0)(max_pool_out_11)
enc2_conv_1_output = layers.Conv1D(filters=32,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_enc_2_1)(enc2_conv_0_output)
max_pool_out_22 = layers.MaxPooling1D(pool_size=2, strides=2, padding="same")(enc2_conv_1_output) 
enc3_conv_0_output = layers.Conv1D(filters=64,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_enc_3_0)(max_pool_out_22)
enc3_conv_1_output = layers.Conv1D(filters=64,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_enc_3_1)(enc3_conv_0_output)
max_pool_out_33 = layers.MaxPooling1D(pool_size=2, strides=2, padding="same")(enc3_conv_1_output)
centr_0 = layers.Conv1D(filters=128,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_cen_0)(max_pool_out_33)
centr_1 = layers.Conv1D(filters=128,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_cen_1)(centr_0) 
dec0_up = layers.UpSampling1D(size=2)(centr_1)
dec_0_conv_up = layers.Conv1D(filters=64,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_0_up)(dec0_up) 
dec0_conc = layers.Concatenate()([enc3_conv_1_output, dec_0_conv_up])
dec_0_conv0 = layers.Conv1D(filters=64,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_0_0)(dec0_conc) 
dec_0_conv1 = layers.Conv1D(filters=64,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_0_1)(dec_0_conv0)

dec1_up = layers.UpSampling1D(size=2)(dec_0_conv1)
dec_1_conv_up = layers.Conv1D(filters=32,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_1_up)(dec1_up) 
dec1_conc = layers.Concatenate()([enc2_conv_1_output, dec_1_conv_up])
dec_1_conv0 = layers.Conv1D(filters=32,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_1_0)(dec1_conc) 
dec_1_conv1 = layers.Conv1D(filters=32,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_1_1)(dec_1_conv0) 

dec2_up = layers.UpSampling1D(size=2)(dec_1_conv1)
dec_2_conv_up = layers.Conv1D(filters=16,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_2_up)(dec2_up) 
dec2_conc = layers.Concatenate()([enc1_conv_1_output, dec_2_conv_up])
dec_2_conv0 = layers.Conv1D(filters=16,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_2_0)(dec2_conc) 
dec_2_conv1 = layers.Conv1D(filters=16,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_2_1)(dec_2_conv0)
 
dec3_up = layers.UpSampling1D(size=2)(dec_2_conv1)
dec_3_conv_up = layers.Conv1D(filters=8,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_3_up)(dec3_up) 
dec3_conc = layers.Concatenate()([second_conv_output, dec_3_conv_up])
dec_3_conv0 = layers.Conv1D(filters=8,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_3_0)(dec3_conc) 
dec_3_conv1 = layers.Conv1D(filters=8,
                                kernel_size=3,
                                strides=1,activation="relu",padding="same",use_bias=False,kernel_initializer=kernel_initializer_dec_3_1)(dec_3_conv0) 

final_layer_out = layers.Conv1D(filters=4,
                                kernel_size=3,
                                strides=1,padding="same",use_bias=False,kernel_initializer=kernel_initializer_final)(dec_3_conv1)
output = layers.Softmax()(final_layer_out)
   

output_npy = output.numpy()
np.savetxt('output.txt', output_npy[0, :, :], fmt='%f')
```
