```python
import matplotlib.pyplot as plt
import numpy as np

x_values = ["#30", "#60", "#90", "#120", "#150", "#169"]
  
original_model = [6.63, 13.26, 19.91, 26.58, 33.29, 37.53]
quantized_model = [3.67, 7.33, 11.01, 14.67, 18.34, 20.66]
  
X_axis = np.arange(len(x_values))
  
bars = plt.bar(X_axis - 0.2, original_model, 0.4, label = 'Original model')
plt.bar_label(bars, padding=3)
bars = plt.bar(X_axis + 0.2, quantized_model, 0.4, label = 'Quantized model')
plt.bar_label(bars, padding=5)

#reduce by halft figure size
plt.gcf().set_size_inches(6, 2)

# Remove the plot borders
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

#set font to times new roman
plt.rcParams["font.family"] = "Times New Roman"

plt.xticks(X_axis, x_values)
plt.xlabel("Number of subjects")
plt.ylabel("Time (s)")
plt.legend()
plt.tight_layout()
plt.gcf().set_dpi(300)
plt.savefig("bar_plot1.svg", format="svg")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

clocks = [338892203, 28664610]
values = ["AIRISC no extensions", "AIRISC SIMD extension"]

plt.rcdefaults()
fig, ax = plt.subplots()

#set fig height
#fig.set_figheight(2.2)

# Example data
y_pos = np.arange(len(clocks))

ax.barh(y_pos, clocks, align='center', linewidth=0)
#plt.xticks(ticks=[100000000,200000000,300000000],fontsize=15, fontname='times new roman')
plt.xticks(ticks=[])

ax.ticklabel_format(style='plain')    # to prevent scientific notation.
ax.set_yticks(y_pos, labels=values)
plt.yticks(y_pos, values, fontsize=15, fontname='times new roman')
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Clock cycles', size=15, fontname='times new roman')
for bars in ax.containers:
    ax.bar_label(bars, fmt='{:,.0f}', label_type='edge', padding=3, size=15, fontname='times new roman')

# Remove the plot borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.spines['left'].set_visible(False)

#set font to times new roman
plt.rcParams["font.family"] = "Times New Roman"
plt.gcf().set_size_inches(6, 1.5)

plt.tight_layout()
plt.gcf().set_dpi(300)
plt.savefig("bar_plot2.svg", format="svg")

plt.show()
```
